# Making Keywords using OpenAI API

```python
import os
# import openai
from openai import OpenAI
OpenAI.api_key=open_ai_key
```

```python
client = OpenAI(api_key = open_ai_key)

def get_prompts(user_prompt):

    completion = client.chat.completions.create(
      model="gpt-4",
      messages=[
        {"role": "system", "content": "Please provide 10 keywords that encapsulate the overall theme, emotion, and atmosphere of the music piece described and its aspect list given. Focus on capturing the essence and mood without mentioning specific instruments or music production details. Include one keyword related to the specific context or setting (like a nightclub, gaming, sports, movie or religious) where the music might be featured. Output only the keywords separated by comma."},
        {"role": "user", "content": user_prompt}
      ]
    )
    # print(completion.choices[0].message)
    return completion.choices[0].message.content
```

```python
import pandas as pd
from concurrent.futures import ProcessPoolExecutor
import concurrent.futures
from tqdm import tqdm
```

```python
caption_keyword_df = caption_df.copy()

with ProcessPoolExecutor() as executor:
    # # Apply the get_prompts function in parallel
    # caption_keyword_df['caption'] = list(executor.map(get_prompts, caption_df['caption']))

    # Create a list of futures
    futures = [executor.submit(get_prompts, prompt) for prompt in caption_df['caption']]
    results = []

    # Iterate over the futures and fill the results list with outputs
    for future in tqdm(concurrent.futures.as_completed(futures), total=len(futures)):
        results.append(future.result())

    caption_keyword_df['caption'] = results
```

# Vector Search Retrieval

```python
client = MongoClient("")
dbName = "MLProj_keywords"
collectionName = "collection_of_keywords_blobs"
collection = client[dbName][collectionName]
```

```python
embeddings = OpenAIEmbeddings(openai_api_key="")
```

```python
vectorStore = MongoDBAtlasVectorSearch( collection, embeddings )
```

```python
def query_data(query):
    # Convert question to vector using OpenAI embeddings
    # Perform Atlas Vector Search using Langchain's vectorStore
    # similarity_search returns MongoDB documents most similar to the query

    docs = vectorStore.similarity_search(query) #, index_name = 'keywordsSearch')
    # print(docs)
    as_outputs = []
    rows = []
    for doc in docs:
        as_output = doc.page_content
        row = doc.metadata['row']
        as_outputs.append(as_output)
        rows.append(row)

    return as_outputs, rows
```

```python
query = "excitement, happy, love"
query_data(query)
```

```python
client = OpenAI(api_key = open_ai_key)

def get_prompts(user_prompt):

    completion = client.chat.completions.create(
      model="gpt-4",
      messages=[
        {"role": "system", "content": "Please provide 10 keywords that capture the thematic elements and emotional essence of the article, suitable for inspiring music composition. The keywords should evoke the main themes and emotions, including one keyword related to the article’s specific subject (like a nightclub, gaming, sports, movie or religious). Avoid nouns, locations, specific or technical terms, focusing on broader themes and the overall mood. Output only the keywords separated by comma."},
        {"role": "user", "content": user_prompt}
      ]
    )

    # print(completion.choices[0].message)
    return completion.choices[0].message.content
```
